```python
## To generate .txt file for autohotkey 

import pandas as pd
import re


df = pd.read_excel("formula.xlsx")  
def elements_from_formula(formula):
    elems = re.findall(r"[A-Z][a-z]?", str(formula))
    elems = sorted(set(elems))
    return elems

queries = {}

for formula in df["Formula"]:
    elems = elements_from_formula(formula)
    key = " AND ".join(elems)
    nelem = len(elems)

    if key not in queries:
        queries[key] = nelem


with open("pcd_batchsearch.txt", "w") as f:
    for q, n in sorted(queries.items()):
        f.write(f"{q} | nelements={n}\n")

print(f"Generated {len(queries)} PCD batch searches.")
```
